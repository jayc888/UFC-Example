

The following are the variables in the dataset and their meanings:

  * `event_id`: Unique identifier for the event.
  * `event_name`: Name of the UFC event.
  * `date`: Date of the event.
  * `location`: Location where the event took place.
  * `fight_id`: Unique identifier for each fight.
  * `division`: Weight division of the fight.
  * `title_fight`: Indicates if it was a title fight (1 for Yes, 0 for No).
  * `method`: Method by which the fight was won (e.g., KO, TKO, Decision, Submission).
  * `finish_round`: The round in which the fight ended.
  * `match_time_sec`: Time in seconds at which the fight ended in the given round.
  * `total_rounds`: Total number of rounds in the fight.
  * `referee`: Name of the referee for the fight.

  * `r_sig_str_atmpted`: Significant strikes attempted by the red corner fighter.
